```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

# Normal Distribution

```python
a = np.random.multivariate_normal([0, 0], [[100, 0], [0, 100]], 150).T
b = np.random.multivariate_normal([100, 0], [[1, 0], [0, 100]], 150).T

plt.plot(a[0], a[1], 'o')
plt.plot(b[0], b[1], 'o')
plt.axis('equal')
plt.title('Original')
plt.show()
```

```python
perplexities = [100]

for perplexity in perplexities:
    operation = TSNE(n_components=2, 
                     perplexity=perplexity,
                     n_iter=5000,
                     n_iter_without_progress=300)

    tsne_a = np.transpose(operation.fit_transform(np.transpose(a)))
    tsne_b = np.transpose(operation.fit_transform(np.transpose(b)))

    plt.plot(tsne_a[0], tsne_a[1], 'o')
    plt.plot(tsne_b[0], tsne_b[1], 'o')
    plt.axis('equal')
    plt.title('Perplexity: %d' % (perplexity))
    plt.show()
```
